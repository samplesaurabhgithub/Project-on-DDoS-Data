```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generate data
data, labels = make_blobs(n_samples=1000, centers=3)
```

```python
# Apply k-means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
# Get cluster labels
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
# Calculate accuracy using DBSCAN evaluation method
accuracy = (cluster_labels == labels).sum() / len(labels)
print('Accuracy:', accuracy)
```

```python
def impute_nan(df,variable,value):
    df[variable+"_mean"]=df[variable].fillna(value)
```
